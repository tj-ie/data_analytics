```python
import pandas as pd
```

```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/data/sample_years.xlsx')
df
```

```python
for i, r in df.iterrows():
  age = r['age']
  age -= 1
  df.iloc[i, 1] = age

df
```

```python
# without for iteration code
df_out = df.iloc[:, 4:].cumsum(axis=1).add(df["age"] - 1, axis=0)
df_out[df.iloc[:, 4:].eq(0)] = 0

df_out = pd.concat([df[["sex", "age", "siblings", "payment"]], df_out], axis=1)
# df_out["age"] -= 1 # if you need to substract 1 from age
print(df_out)
```

```python
for index, row in df.iterrows():
  age = int(row['age'])
  for col_idx, col_value in enumerate(row[4:]):
    if col_value == 1:
      age += 1
      df.iloc[index, col_idx + 4] = age

df
```

```python
writer = pd.ExcelWriter('/content/drive/MyDrive/Colab Notebooks/data/processed_sample_years.xlsx')
df.to_excel(writer, sheet_name='2012-2021', index=False)

l_years = list(range(2012, 2022))

for year in l_years:
  df.index[df[year] != 0].tolist()
  each_df = df.iloc[df.index[df[year] != 0].tolist()]
  
  each_df.to_excel(writer, sheet_name=f'{year}', index=False)

writer.save()
writer.close()
```
